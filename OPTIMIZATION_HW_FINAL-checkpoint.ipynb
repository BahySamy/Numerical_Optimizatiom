```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,20,50)
y = -2 * x +1
x = x.reshape(x.shape[0],1)
ones = np.ones(x.shape[0]).reshape((x.shape[0], 1))
x = np.hstack([ones,x])
```

```python
plt.scatter(x[:,1],y)
```

```python
# Adagrad

def Adagrad(x, y, alpha, epsoln, number_of_iterations):
   
    theta = np.random.random(2)
    cost_array =[0]
    theta_array =[[0,0]]
    count = 0
    vt_1 = 0

    for i in range(number_of_iterations):
        
        h_theta = x @ theta
        gradient = (1/len(y))*(h_theta - y ) @ x
        vt = vt_1 + (gradient ** 2)
        theta = theta - alpha / (np.sqrt(vt)+epsoln)  * gradient
        cost = (1/len(y)) * sum((h_theta - y)**2)
        
        theta_array.append(theta)
        cost_array.append(cost)
        
        count += 1
        
        if np.linalg.norm(gradient,2) < 1e-3 :
            break
        
        if abs(cost_array[-1] - cost_array[-2]) < 1e-6 :
            break

    y_pred = x @ theta
    
    return theta_array, cost_array, y_pred, count
```

```python
theta_array, cost_array, y_pred, count = Adagrad(x, y, .001, 1e-8, 10000)
```

```python
print(f'Number of iterations: {count}')
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

```python
plt.plot(np.arange(count), cost_array[1:])
plt.title('Epocks Vs cost Function')
plt.xlabel('Numbers of Epochs')
plt.ylabel('Cost Function')
```

```python
weights = np.array(theta_array)
print(weights.shape)
plt.plot(weights[: , 0][1:], cost_array[1:] , label = 'theta0')
plt.plot(weights[: , 1][1:], cost_array[1:], label = 'theta1')
plt.title('Theta Vs cost Function')
plt.xlabel('Theta')
plt.ylabel('Cost Function') 
plt.legend()
```

```python
plt.scatter(x[:,1], y)
plt.plot(x[:,1], y_pred, 'r')
plt.title('Real Output Vs Predicted Output')
plt.xlabel('X Values')
plt.ylabel('Y Values')
```

```python
# Adagrad

def RMSProp(x, y, alpha, epsoln, beta,number_of_iterations):
    
    
    theta = np.random.random(2)
    cost_array =[0]
    theta_array =[[0,0]]
    count = 0
    vt_1 = 0

    for i in range(number_of_iterations):
        
        h_theta = x @ theta
        gradient = (1/len(y))*(h_theta - y ) @ x
        vt = (beta * vt_1) + (1-beta)*(gradient ** 2)
        theta = theta - alpha / (np.sqrt(vt)+epsoln)  * gradient
        cost = (1/len(y)) * sum((h_theta - y)**2)
        
        theta_array.append(theta)
        cost_array.append(cost)
        
        count += 1
        
        if np.linalg.norm(gradient,2) < 1e-3 :
            break
        
        if abs(cost_array[-1] - cost_array[-2]) < 1e-6 :
            break

    y_pred = x @ theta
    
    return theta_array, cost_array, y_pred, count
```

```python
theta_array, cost_array, y_pred, count = RMSProp(x, y, .0001, 1e-8, .92, 14000)
```

```python
print(f'Number of iterations: {count}')
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

```python
plt.plot(np.arange(count), cost_array[1:])
plt.title('Epocks Vs cost Function')
plt.xlabel('Numbers of Epochs')
plt.ylabel('Cost Function')
```

```python
# Adam

def Adam(x, y, alpha, epsoln, beta, gamma, number_of_iterations):
    
    
    theta = np.random.random(2)
    cost_array =[0]
    theta_array =[[0,0]]
    count = 0
    vt_1 = 0
    mt_1 = 0

    for i in range(number_of_iterations):
        
        h_theta = x @ theta
        gradient = (1/len(y))*(h_theta - y ) @ x
        vt = (beta * vt_1) + (1-beta)*(gradient ** 2)
        mt = gamma * mt_1 + (1-gamma) * gradient
        
        vt = vt / (1-beta ** (i+1))
        mt = mt / (1- gamma ** (i+1))
        
        theta = theta - alpha / (np.sqrt(vt)+epsoln)  * mt
        cost = (1/len(y)) * sum((h_theta - y)**2)
        
        theta_array.append(theta)
        cost_array.append(cost)
        
        count += 1
        
        if np.linalg.norm(gradient,2) < 1e-3 :
            break
        
        if abs(cost_array[-1] - cost_array[-2]) < 1e-6 :
            break

    y_pred = x @ theta
    
    return theta_array, cost_array, y_pred, count
```

```python
theta_array, cost_array, y_pred, count = Adam(x, y, .001, 1e-8, .999, .9, 15000)
```

```python
print(f'Number of iterations: {count}')
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

```python
plt.plot(np.arange(count), cost_array[1:])
plt.title('Epocks Vs cost Function')
plt.xlabel('Numbers of Epochs')
plt.ylabel('Cost Function')
```
