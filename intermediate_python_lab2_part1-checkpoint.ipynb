```python
# importing needed libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# Question 1 

```python
# importing img
img = mpimg.imread('Desktop/dog.png')
# Model Output
model_output = np.array([4.1542e-04, 9.9958e-01])
```

```python
# plotting 

# initializing figure and supplots
my_fig = plt.figure("Testing My Model") 
sub1 = my_fig.add_subplot(121)
sub2 = my_fig.add_subplot(122)

# plotting img
sub1.imshow(img)
sub1.set_title("Testing Image")
sub1.axis('off')

# plotting the propability 
sub2.barh(['Cat', 'Dog'], model_output)
sub2.set_title("Model Output Probablity")



plt.tight_layout()
```

# Question 2

```python
# setting parameters 
men_values = np.array([20, 35, 30, 35, 27])
women_values = np.array([25, 32, 34, 20, 25])

ind = np.arange(5)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
menBar = ax.bar(ind, men_values, width, color='g')

womenBar = ax.bar(ind, women_values, bottom = men_values,width=width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((menBar, womenBar), ('Men', 'Women'))


plt.show()
```
