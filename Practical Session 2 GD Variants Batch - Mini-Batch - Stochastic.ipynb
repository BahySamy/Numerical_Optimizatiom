#### Import numpy, matplotlib.pyplot and make it inline

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

To have a dataset or set of data points, the student must generate a pair of arrays <b>X</b> and <b>y</b> with the values in <b>X</b> equally distributed between <b>0</b> and <b>20</b> and the values in <b>y</b> such that: 
<b>yi = a*xi + b (and a = -2, b = 1)</b>


```python
x = np.linspace(0,20,50)
y = -2 * x +1
x = x.reshape(x.shape[0],1)
ones = np.ones(x.shape[0]).reshape((x.shape[0], 1))
x = np.hstack([ones,x])
```

#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python


def Batch(x, y, alpha, number_of_iterations):
    
    
    theta = np.random.random(2)
    cost_array =[0]
    theta_array =[[0,0]]
    count = 0

    for i in range(number_of_iterations):
        
        h_theta = x @ theta
        gradient = (1/len(y))*(h_theta - y ) @ x
        theta = theta - alpha * gradient
        cost = (1/len(y)) * sum((h_theta - y)**2)
        
        theta_array.append(theta)
        cost_array.append(cost)
        
        count += 1
        
        if np.linalg.norm(gradient,2) < 1e-3 :
            break
        
        if abs(cost_array[-1] - cost_array[-2]) < 1e-6 :
            break

    y_pred = x @ theta
    
    return theta_array, cost_array, y_pred, count
```

#### Try your function with any input parameterss of your choice.

```python
theta_array, cost_array, y_pred, count = Batch(x, y, .001, 10000)
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def Mini_Batch(x, y, alpha, number_of_iterations, Batch_size):
    
    
    theta = np.random.random(2)
    cost_array =[0]
    theta_array =[[0,0]]
    count = 0
    
    for k in range(number_of_iterations):
        for i in range(int(x.shape[0]/ Batch_size)):
            x_batch = x[i * Batch_size : (i+1) * Batch_size]
            y_batch = y[i * Batch_size : (i+1) * Batch_size]

            h_theta = x_batch @ theta
            gradient = (1/len(y_batch))*(h_theta - y_batch) @ x_batch
            theta = theta - alpha * gradient
        
        cost = (1/len(y_batch)) * sum((h_theta - y_batch)**2)
        theta_array.append(theta)
        cost_array.append(cost)
        count += 1
            
        if np.linalg.norm(gradient,2) < 1e-2 :
            break
            
        if abs(cost_array[-1] - cost_array[-2]) < 1e-6 :
            break

    y_pred = x @ theta
    
    return theta_array, cost_array, y_pred, count
```

```python
theta_array, cost_array, y_pred, count = Mini_Batch(x, y, .001, 10000, 10)
```

```python
print(f'Number of iterations: {count}')
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)
```

## Stochastic GD

#### Update the implementation of the Batch GD to beStochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to Batch and Mini-Batch GD

```python
def stochastic(x, y, alpha, number_of_iterations):
    
    theta = np.random.random(2)
    cost_array =[0]
    theta_array =[[0,0]]
    count = 0

    for k in range(number_of_iterations):
        for i in range(x.shape[0]):
            h_theta = theta @ x[i] 
            gradient = (h_theta - y[i]) * x[i]
            theta = theta - alpha * gradient
            
        cost = (h_theta - y[i])**2
        theta_array.append(theta)
        cost_array.append(cost)
                          
        count += 1
        if np.linalg.norm(gradient,2) < 1e-2 :
            break  
                          
        if abs(cost_array[-1] - cost_array[-2]) < 1e-6 :
            break
                          
    y_pred = x @ theta
    return theta_array, cost_array, y_pred, count
```

```python
theta_array, cost_array, y_pred, count = stochastic(x, y, .001, 10000)
```
