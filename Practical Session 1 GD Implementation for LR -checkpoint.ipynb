## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate. 

```python
theta = np.array([0,0])
alpha = .003
print(theta)
print(alpha)
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
#### Note: you will need to iterate through all data points

```python
h_theta = theta[0] + theta[1] * x
print(h_theta)
```

### Step3: Calculate Cost function (theta_0,theta_1 ).


```python
j_function = (1/(2*13))*sum((h_theta - y)**2)
print(j_function)
```

### Step4: Calculate the gradient.


```python
gradient0 = (1/13) * sum(h_theta - y)
gradient1 = (1/13) * sum((h_theta - y) * x)
gradient = np.array([gradient0, gradient1])
print(gradient)
```

### Step5: Update the parameters (simultaneously).


```python
theta = theta - alpha * gradient
print(theta)
```

### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.

```python
count = 0
j_function = (1/(2*13))*sum((h_theta - y)**2)
while np.linalg.norm(gradient,2) > 1e-5:
    theta= theta - (alpha * gradient)
    h_theta = theta[0] + theta[1] * x
    gradient0 = (1/13) * sum(h_theta - y)
    gradient1 = (1/13) * sum((h_theta - y) * x)
    gradient = np.array([gradient0, gradient1])
    count += 1
    
print(np.linalg.norm(gradient,2))
print(f'number of eteration: {count}')
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x

```python
h_theta = theta[0] + theta[1] * x
print(h_theta)
```

#### Plot  LR equation output (fitted line) with the original data (scatter plot of X,y)

```python
plt.scatter(x,y)
plt.plot(x,h_theta)
plt.title('Real Output Vs Predicted Output')
plt.xlabel('X Values')
plt.ylabel('Y Values')
```

#### Use R2 score to evaluate LR equation output



https://en.wikipedia.org/wiki/Coefficient_of_determination

```python
from sklearn.metrics import r2_score
r2_score(y, h_theta)
```
