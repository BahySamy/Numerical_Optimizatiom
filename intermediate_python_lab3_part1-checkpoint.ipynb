# Question 2


```python
# importing data
tit_data1 = pd.read_csv('Desktop/LAB_files/2 process_Titanic_datasets/titanic-passengers.csv', sep = ';')
tit_data2 = pd.read_json('Desktop/LAB_files/2 process_Titanic_datasets/titanic-passengers.json')
```

```python
tit_data1.head()
```

```python
tit_data2.head()
```

```python
# transforming field into dataframe
fields = tit_data2['fields']
titanic_data2 = pd.json_normalize(fields)
titanic_data2.head()
```

```python
#removing some column from titanic data1 
tit_data1.drop(['Ticket', 'Age'], inplace = True,axis = 1)
tit_data1.head()
```

```python
#removing some column from titanic data2
titanic_data2.drop(['fare', 'cabin'], inplace = True,axis = 1)
titanic_data2.head()
```

```python
# merging 2 data sets
final_titanic_data = pd.merge(tit_data1,titanic_data2, left_on = 'PassengerId' , right_on = 'passengerid')
final_titanic_data.head()
```

```python
final_titanic_data.columns
```

```python
# Drop repeated columns 
finaly =final_titanic_data.drop(['passengerid','name','embarked','sex', 'survived', 'sibsp','parch', 'pclass'], axis = 1)
```

```python
finaly.shape
```

```python
# saving final data as csv 
finaly.to_csv('Desktop/LAB_files/2 process_Titanic_datasets/final_data.csv',index = False)
```
